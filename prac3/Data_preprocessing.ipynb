```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
```

```python
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
print("\nMissing values :\n",df.isnull().sum())
```

```python
df.loc[5:10, 'crim'] = np.nan
df.loc[15:18, 'age'] = np.nan
df
```

```python
#impute missing values
df['crim']=df['crim'].fillna(df['crim'].mean())
df['age']=df['age'].fillna(df['age'].median())
df
```

```python
#detect outliers using IQR
q1 = df['tax'].quantile(0.25)
q3 = df['tax'].quantile(0.75)
IQR = q3-q1
print(IQR)
outlier =df[(df['tax'] <(q1-1.5*IQR)) | (df['tax'] >(q3+1.5*IQR))]
print(f"Number of outliers in 'tax': {outlier.shape[0]}")
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(df['tax'])
plt.title("Boxplot of tax")
plt.show()
```
