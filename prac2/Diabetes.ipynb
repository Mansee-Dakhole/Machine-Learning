```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
```

```python
#load the dataset
diabetes = load_diabetes()
df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
```

```python
#show the first 5 rows of data
df.head()
```

```python
# show the information about dataset
df.info()
```

```python
#gives the statistical summary of data
df.describe()
```

```python
# Count and display the number of occurrences of each unique value in the 'target' column
print(df['target'].value_counts())
```

```python
# Plot histograms for all numeric columns with custom size and bin count
df.hist(figsize = (10,6), bins = 20)
plt.suptitle("Histogram for Feature Columns")
plt.show()
```

```python
# Pair plot of features colored by species
sns.pairplot(df, hue='target')
plt.suptitle("pair plot of Diabetes fratures",y=1.02)
plt.show()
```

```python
# Horizontal boxplot of all features
plt.figure(figsize=(8,6))
sns.boxplot(data = df, orient="h")
plt.title("Boxplot of all Features")
plt.show()
```

```python
# Heatmap of feature correlations
plt.figure(figsize=(8,6))
sns.heatmap(df.iloc[:,:-1].corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
